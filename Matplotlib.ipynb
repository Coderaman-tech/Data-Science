```python
val=[1,9,3,2,10]#value
lab=['o','b','c','q','l']#label
col=['green','yellow','red','black','blue']#colour
plt.pie(val,labels=lab,autopct='%1.1f%%',colors=col)
```

```python
#to divide some part in pie chart
val=[1,9,3,2,10]#value
lab=['o','b','c','q','l']#label
col=['green','yellow','red','black','blue']#colour
plt.pie(val,labels=lab,autopct='%1.1f%%',colors=col,explode=[0.5,0.4,0,0,0])
```

```python
#to remove unwanted written part in output use plt.show()
val=[1,9,3,2,10]#value
lab=['o','b','c','q','l']#label
col=['green','yellow','red','black','blue']#colour
plt.pie(val,labels=lab,autopct='%1.1f%%',colors=col)
plt.show()
```
